```python
import torch
import matplotlib.pyplot as plt
```

```python
# FM DA
fm_da_path = 'FM_DA/results/metrics'

# FM CORESET METHOD
fm_da_10 = torch.load(f"{fm_da_path}/test_accuracies_DA_10.pth", map_location=torch.device('cpu'))[-1]
fm_da_05 = torch.load(f"{fm_da_path}/test_accuracies_DA_05.pth", map_location=torch.device('cpu'))[-1]
fm_da_01 = torch.load(f"{fm_da_path}/test_accuracies_DA_01.pth", map_location=torch.device('cpu'))[-1]

fm_da_values = [fm_da_10, fm_da_05, fm_da_01]

# FM KMEANS METHOD
fm_da_kmeans_10 = torch.load(f"{fm_da_path}/test_accuracies_DA_10_KMEANS.pth", map_location=torch.device('cpu'))[-1]
fm_da_kmeans_05 = torch.load(f"{fm_da_path}/test_accuracies_DA_05_KMEANS.pth", map_location=torch.device('cpu'))[-1]
fm_da_kmeans_01 = torch.load(f"{fm_da_path}/test_accuracies_DA_01_KMEANS.pth", map_location=torch.device('cpu'))[-1]

fm_da_kmeans_values = [fm_da_kmeans_10, fm_da_kmeans_05, fm_da_kmeans_01]

# FM CORESET METHOD
fm_da_coreset_10 = torch.load(f"{fm_da_path}/test_accuracies_DA_10_CORESET.pth", map_location=torch.device('cpu'))[-1]
fm_da_coreset_05 = torch.load(f"{fm_da_path}/test_accuracies_DA_05_CORESET.pth", map_location=torch.device('cpu'))[-1]
fm_da_coreset_01 = torch.load(f"{fm_da_path}/test_accuracies_DA_01_CORESET.pth", map_location=torch.device('cpu'))[-1]

fm_da_coreset_values = [fm_da_coreset_10, fm_da_coreset_05, fm_da_coreset_01]
```

```python
# plot all values in a single plot with labels
plt.figure(figsize=(10, 5))
plt.plot([10, 5, 1], sup_values, label='SUP')
plt.plot([10, 5, 1], sup_coreset_values, label='SUP CORESET')
plt.plot([10, 5, 1], fm_values, label='FM')
plt.plot([10, 5, 1], fm_kmeans_values, label='FM KMEANS')
plt.plot([10, 5, 1], fm_coreset_values, label='FM CORESET')
plt.plot([10, 5, 1], fm_da_values, label='FM DA')
plt.plot([10, 5, 1], fm_da_kmeans_values, label='FM DA KMEANS')
plt.plot([10, 5, 1], fm_da_coreset_values, label='FM DA CORESET')
plt.xlabel('Percentage of labeled data')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Percentage of labeled data')
plt.legend()
plt.show()
```
